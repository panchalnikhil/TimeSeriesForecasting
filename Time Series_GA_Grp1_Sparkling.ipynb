# Check for stationarity on Time Series data (Full data & Train data).

What are characteristics of stationarity:

- __Mean__ should be __constant__

- __Variance__ should be __constant__

- __Covariance__ for a time period should be __constant__

__Hypothesis to check Stationarity:__

 __H0__ : Time series is __non-stationary__
    
 __H1__ : Time series is __stationary__

__Note:__ If the series is non-stationary, stationarise the time series by taking a difference of the time series. Then we can use this particular differenced series to train the models. __We do not need to worry about stationarity for the Test Data__ because we are not building any models on the test Data, we are only evaluating our models over there.

__Auto-Correlation Plot:__

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df)
plt.show();
```

__Insights:__

We can quantify the strength and type of relationship between observations and their lags. In statistics, this is called correlation, and when calculated against lag values in time series it is called autocorrelation (self-correlation).

A correlation value calculated between two groups of numbers, such as observations and their lag1 values, results in a number between -1 and 1. The sign of this number indicates a negative or positive correlation respectively. 
A value close to zero suggests a weak correlation, whereas a value closer to -1 or 1 indicates a strong correlation.
Dotted lines are provided that indicate any correlation values above those lines are statistically significant (meaningful).

## ACF & PACF

A nonseasonal ARIMA model is classified as an "ARIMA(p,d,q)" model, where:

- p is the number of autoregressive terms,
- d is the number of nonseasonal differences needed for stationarity, and
- q is the number of lagged forecast errors in the prediction equation.

Now, the question is how to find out these values based on ACF and PACF graph.

### Original Series

```python
# ACF and PACF with original time series
fig, axes = plt.subplots(1, 2)
fig.set_figwidth(12)
fig.set_figheight(4)
smt.graphics.plot_acf(train, lags=50, ax=axes[0],
                      title='Sparkling Wine: Autocorrelation')
smt.graphics.plot_pacf(train, lags=50, ax=axes[1],
                       title='Sparkling Wine: Partial Autocorrelation')
plt.tight_layout()
```

Here, we could see very less PACF and ACF has significant correlation at lag 1 but after that it is decaying and becoming insignificant.

Hence, try differencing the order using d = 1 and try to evaluate ACF and PACF again.

### Differenced Time series with (d = 1)

```python
# ACF with order difference as "1" as the series was not stationary (refer stationary point#5)
fig, axes = plt.subplots(1, 2)
fig.set_figwidth(12)
fig.set_figheight(4)
smt.graphics.plot_acf(train.diff().dropna(), lags=15, ax=axes[0],
                      title='Sparkling Wine: Differenced Series Autocorrelation (ACF)')
smt.graphics.plot_pacf(train.diff().dropna(), lags=15, ax=axes[1],
                       title='Sparkling Wine: Differenced Series Partial Autocorrelation (PACF)')
plt.tight_layout()
```
